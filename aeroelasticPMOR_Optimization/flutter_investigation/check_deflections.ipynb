# Check deflections

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import os
import pandas as pd
import pickle
```

```python
# Gravity off data
# Open the pickle file and get the wing deflection without running the code
cwd = os.getcwd()
targetpath = cwd+'/aoa_wingDef_investigation_gravityOff'
filepath = targetpath+'/gravityOffData.pickle'
infile = open(filepath, 'rb')
dataGoff = pickle.load(infile)
infile.close()
# Gravity on data
# Open the pickle file and get the wing deflection without running the code
cwd = os.getcwd()
targetpath = cwd+'/aoa_wingDef_investigation'
filepath = targetpath+'/gravityOnData.pickle'
infile = open(filepath, 'rb')
data = pickle.load(infile)
infile.close()
```

```python
fig,ax = plt.subplots()
ax.plot(data['y'][:,0],data['z'][:,0],'b-x')
ax.plot(data['y'][:,4],data['z'][:,4],'r-o')
ax.plot(data['y'][:,8],data['z'][:,8],'g-^')

ax.plot(dataGoff['y'][:,0],dataGoff['z'][:,0],'b--x')
ax.plot(dataGoff['y'][:,4],dataGoff['z'][:,4],'r--o')
ax.plot(dataGoff['y'][:,8],dataGoff['z'][:,8],'g--^')


ax.legend(["AoA=0.0deg","AoA=1.2deg","AoA=2.4deg",
          "AoA=0.0deg Gravity Off","AoA=1.2deg Gravity Off",
           "AoA=2.4deg Gravity Off"])
ax.grid(True)
plt.axis('equal')
ax.set_ylabel('y (m)')
ax.set_xlabel('z (m)')
plt.show()
```
